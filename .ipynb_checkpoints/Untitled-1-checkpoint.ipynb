```python
word_freq = {}
```

```python
path =  "D:\\Projectach\\信息检索_网络小说检索分类\\玄幻"
mkdir = os.listdir(path)
for fi in mkdir:
    novelName = fi

    orinovelPath = "D:\\Projectach\\信息检索_网络小说检索分类"+"\\玄幻\\"+novelName
    file_handle=open(orinovelPath,mode='r',encoding='utf-8')
    content=file_handle.read()
    print(novelName,"小说字数:",len(content))
    words = pseg.cut(content) #jieba默认模式
    l = []
    for word, flag in words:
        if  word not in stop and word not in punc:
            l.append(word)

    cou = Counter(l).most_common(100000)
    for word_pair in cou: 
        try:
            word_freq[word_pair[0]] += 1
        except:
            word_freq[word_pair[0]] = 1
    
```

```python
resPairs = sorted(word_freq.items(),key = lambda x:x[1],reverse = True)
```

```python
print(resPairs)
```

```python
f=open('玄幻词频.txt',mode='w+',encoding='utf-8')

for resPair in resPairs:
    f.write(str(resPair) + os.linesep)
    # print(co,end="\n")
```
